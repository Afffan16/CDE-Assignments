- Used Selenium driver to get the webpage 
- Scrolled down using driver.execute_script to load all the data on the page
- Extracted multiple tables from market summary page of psx 
- Used Pandas Excel writer to write each table on a different sheet all in one excel file
- Used thead - tr for naming the sheets on excel for each table
- from tbody of each table skipped the first tr ( contains headers )
- from trs in tbody got all tds and saved the text of tds in variables
- passed the variables in a list data[]
- appended the list data[] in table_data[]
- used for loop for all the trs in a table
- used another for loop for all the tables on the page





```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
import time
import re

options = Options()
# options.add_argument("--headless") 
path = r'D:\@ff@n\Cloud Data Engineering\Selenium Master Project\Selenium-MasterProject\chromedriver-win64\chromedriver.exe'
driver = webdriver.Chrome(path)

driver.get("https://www.psx.com.pk/market-summary/")
time.sleep(3)

scroll_height = driver.execute_script("return document.body.scrollHeight")
current_position = 0
while current_position < scroll_height:
    current_position += 300
    driver.execute_script(f"window.scrollTo(0, {current_position});")
    time.sleep(0.5)
    scroll_height = driver.execute_script("return document.body.scrollHeight")
time.sleep(2)

containers = driver.find_elements(By.CLASS_NAME, 'table-responsive')
containers = containers[2:-3]

columns = ['Scrip', 'LDPC', 'OPEN', 'HIGH', 'LOW', 'Current', 'Change', 'Volume']
excel_path = 'PSX_MarketData.xlsx'
writer = pd.ExcelWriter(excel_path, engine='xlsxwriter')

for tblno, container in enumerate(containers):
    try:
        table = container.find_element(By.TAG_NAME, 'table')
        thead = table.find_element(By.TAG_NAME, 'thead')
        header = thead.find_element(By.TAG_NAME, 'tr').text.strip()

        sheet_name = re.sub(r'[\[\]\:\*\?\/\\]', '', header)
        sheet_name = sheet_name[:31] if sheet_name else f'Table_{tblno+1}'

        tbody = table.find_element(By.TAG_NAME, 'tbody')
        table_rows = tbody.find_elements(By.TAG_NAME, 'tr')[1:]

        table_data = []
        for table_row in table_rows:
            tds = table_row.find_elements(By.TAG_NAME, 'td')
  
            scrip = tds[0].text
            ldpc = tds[1].text
            open = tds[2].text
            high = tds[3].text
            low = tds[4].text 
            current = tds[5].text
            change = tds[6].text
            volume = tds[7].text

            data = [scrip, ldpc, open, high, low, current, change, volume]
            table_data.append(data)

        df = pd.DataFrame(table_data, columns=columns)
        df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"Added table: {sheet_name} with {len(table_data)} rows")

    except Exception as e:
        print(f"Failed to process table {tblno+1}: {e}")

writer.close()
driver.quit()
print("Data saved to PSX_MarketData.xlsx")
```
